```python
#############
# задание 1 #
#############

# перепешите последний пример
# так, чтобы в нём появилась
# обработка правила "2",
# рисующая контурный круг

# подсказки:
# пустая заливка — fill(None)
# цвет обводки — stroke(0, 0, 0)
# толщина — strokeWidth(3)

from drawbot_skia.drawbot import *

newDrawing()

rules = [2, 1, 0, 1, 0, 2, 0, 1, "гусь", 2, 2] * 20

w, h = 742.5, 1050
newPage(w, h)

margin = 80
x = margin
y = h - margin

step = (w - margin * 2) / 6

size = step

for rule in rules:
    if rule == 0:
        fill(200/255, 0/255, 20/255)
        stroke(None)
        oval(x, y - step, size, size)
    elif rule == 1:
        fill(100/255, 100/255, 0/255)
        stroke(None)
        rect(x, y - step, size, size)
    elif rule == 2:
        fill(None)
        stroke(0/255, 100/255, 100/255)
        strokeWidth(3)
        oval(x, y - step, size, size)
    else:
        print(rule, "<— неизвестное правило")

    x += step

    if x >= w - margin:
        x = margin
        y -= step

    if y - step < margin:
        newPage(w, h)
        x = margin
        y = h - margin

saveImage("task1pic.pdf")
```

```python
##########################
# задание 2              #
# творческое, свободное  #
##########################

# можете использовать результат
# задания 1 как основу, или
# попробовать написать что-то
# новое

from drawbot_skia.drawbot import *

newDrawing()

rules = [2, 3, 1, 0, 1, 3, 0, 2, 3, 3, 2, 0, 1, "гусь", 3, 2, 2] * 20

w, h = 742.5, 1050
newPage(w, h)

margin = 80
x = margin
y = h - margin

step = (w - margin * 2) / 6

size = step

for rule in rules:
    if rule == 0:
        fill(200/255, 0/255, 20/255)
        stroke(None)
        oval(x, y - step, size, size)
    elif rule == 1:
        fill(100/255, 100/255, 0/255)
        stroke(None)
        rect(x, y - step, size, size)
    elif rule == 2:
        fill(None)
        stroke(0/255, 100/255, 100/255)
        strokeWidth(3)
        oval(x, y - step, size, size)
    elif rule == 3:
        fill(None)
        stroke(150/255, 0/255, 150/255)
        strokeWidth(5)
        line((x, y - step), (x + size, y - step))
    else:
        print(rule, "<— неизвестное правило")

    x += step

    if x >= w - margin:
        x = margin
        y -= step

    if y - step < margin:
        newPage(w, h)
        x = margin
        y = h - margin

saveImage("task2pic.pdf")
```
